# 1.4.3 Geometric Transformations of Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('img3.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image', img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image', img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Translation

```python
img = cv2.imread('img4.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,150]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rotation

```python
img = cv2.imread('img3.jpg',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),40,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Affine Transformation

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[235,457],[238,269],[569,231],[564,493]])
pts2 = np.float32([[235,457],[235,270],[570,270],[570,457]])
#M = cv2.getAffineTransform(pts1,pts2)
M = cv2.getPerspectiveTransform(pts1,pts2)
#dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.warpPerspective(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('simg',img)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('dsu5.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[579,338],[753,377],[585,532],[762,541]])
pts2 = np.float32([[450,300],[700,300],[450,500],[700,500]])
#M = cv2.getAffineTransform(pts1,pts2)
M = cv2.getPerspectiveTransform(pts1,pts2)
#dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.warpPerspective(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('simg',img)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
